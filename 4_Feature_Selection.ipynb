```python
# dataset =pd.read_csv('/content/drive/MyDrive/master-iot-data/cleaned_merged_data.csv.csv')
dataset =pd.read_csv('/content/drive/MyDrive/master-iot-data/cleaned_merged_data_MinMax.csv')
data = dataset.copy()
data
```

```python
threshold = 0.01  # Eşik değeri belirleyin

# Varyansı hesaplayın
variances = data.var()

# Eşik değerden düşük olan sütunları seçin
selected_features = variances[variances > threshold].index

# Seçilen özellikleri görüntüleyin
print("Selected Features:")
print(selected_features)

# Seçilen özellikleri içeren yeni bir veri kümesi oluşturun
selected_data = data[selected_features]
```

```python
data = selected_data.copy()
```

```python
# Tüm özelliklerin adlarını alın
all_features = data.columns

# Seçilen özellikleri içeren listeden tüm özelliklerin adlarını çıkarın
not_selected_features = set(all_features) - set(selected_features)

# Seçilmeyen özellikleri görüntüleyin
print("Not Selected Features:")
print(not_selected_features)
```
